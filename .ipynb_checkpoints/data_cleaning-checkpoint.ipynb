```python
df['mileagekm'] = df['mileagekm'].str.replace(',', '')
```

```python
df["mileagekm"] = df["mileagekm"].apply(pd.to_numeric)
```

```python
df["enginecapacitycc"] = df["enginecapacitycc"].apply(pd.to_numeric)
df["pricePKR"] = df["pricePKR"].apply(pd.to_numeric)
```

```python
df.dtypes
```

```python
df.enginecapacitycc.fillna(df['enginecapacitycc'].mean(),inplace=True)
```

```python
df.pricePKR.fillna(df['pricePKR'].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
df.fueltype.fillna('CompressedNaturalGas',inplace=True)
```

```python
df.fueltype.value_counts()
```

```python
df.fillna({
        'location': 'unknownLocation',
        'color': 'unknownColor',
        'assembly': 'unknownAssembly',
        'automatic': 'unknown',
        
    }, inplace= True)
```

```python
df.isna().sum()
```
